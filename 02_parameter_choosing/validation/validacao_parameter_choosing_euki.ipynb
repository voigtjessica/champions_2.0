# Validation tm - parameter choosing

The goal of this notebook it's to verify if there is significative difference between models performed from the same corporea with a fixed seed, established by the UMAP model.
I here perform tests with results from parameter choosing dfs that used EU Data and AI documents. 

```python
import csv
import pandas as pd
```

```python
def compare_csv_files(file1, file2):
    with open(file1, 'r') as f1, open(file2, 'r') as f2:
        csv1 = csv.reader(f1)
        csv2 = csv.reader(f2)

        for row1, row2 in zip(csv1, csv2):
            if row1 != row2:
                return False

    return True

file1 = 'C:\\Users\\opc\\Documents\\Python_Scripts\\champions_2.0\\02_parameter_choosing\\df_examples_models_daten_ki_eu.csv'
file2 = 'C:\\Users\\opc\\Documents\\Python_Scripts\\champions_2.0\\02_parameter_choosing\\validacao_df_examples_models_daten_ki_eu.csv'

if compare_csv_files(file1, file2):
    print('Os arquivos são idênticos')
else:
    print('Os arquivos não são idênticos')
```

```python
original = pd.read_csv('C:\\Users\\opc\\Documents\\Python_Scripts\\champions_2.0\\02_parameter_choosing\\df_examples_models_daten_ki_eu.csv', sep=";")
segundo = pd.read_csv('C:\\Users\\opc\\Documents\\Python_Scripts\\champions_2.0\\02_parameter_choosing\\validacao_df_examples_models_daten_ki_eu.csv', sep=";")
```

Vamos verificar se os arquivos sao iguais e apenas a distribuicao do número dos modelos foi diferente

```python
# comparando quantidades de tópicos por modelo

def concatenate_grouped_dataframes(df1, df2, group_by_column):
    # agrupar o primeiro DataFrame por group_by_column
    grouped1 = df1.groupby([group_by_column]).max('Topic')

    # agrupar o segundo DataFrame por group_by_column
    grouped2 = df2.groupby([group_by_column]).max('Topic')

    # concatenar os resultados de groupby lado a lado
    result = pd.concat([grouped1, grouped2], axis=1)

    return result

# 
group_by_column = 'model'

result = concatenate_grouped_dataframes(original, segundo, group_by_column)
result
```

Os modelos definitivamente sao diferentes, apesar de eu ter gerado random state iguais!

```python
def are_identical_dataframes_ignore_order(df1, df2):
    # ordenar os DataFrames antes de compará-los
    df1 = df1.sort_values(by=df1.columns.tolist()).reset_index(drop=True)
    df2 = df2.sort_values(by=df2.columns.tolist()).reset_index(drop=True)

    # comparar os valores dos DataFrames
    if not df1.equals(df2):
        return False

    return True


if are_identical_dataframes_ignore_order(segundo, terceiro):
    print('Os dois DataFrames são idênticos')
else:
    print('Os dois DataFrames não são idênticos')
```
